```python
## Table Households
mycursor = mydb.cursor()
sql = "INSERT INTO Households(hh_id,hh_race,hh_is_latinx,hh_zip_code,hh_state,hh_income,hh_size,hh_residence_type) VALUES (%s,%s,%s,%s,%s,%s,%s,%s)"

n = 0
for df_chunk in pd.read_csv("~/Downloads/dta_at_hh.csv", chunksize = 10000):
    del df_chunk['Unnamed: 0']
    n1 = df_chunk.shape[0]
    n = n+n1
    
    tuple_value = list(df_chunk.itertuples(index=False, name = None))
    for i in tuple_value:
        val = i
        mycursor.execute(sql, val)
    print(n)

mydb.commit()
mycursor.close()
```
